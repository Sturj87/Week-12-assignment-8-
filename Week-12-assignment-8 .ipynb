```python
plt.figure(figsize=(8, 6))
df['class'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8, 6))
df['odor'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Distribution of Odors')
plt.xlabel('Odor')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8, 6))
df['habitat'].value_counts().plot(kind='bar', color='salmon')
plt.title('Distribution of Habitats')
plt.xlabel('Habitat')
plt.ylabel('Count')
plt.show()


```

```python
colors = ['blue' if x == 'e' else 'orange' for x in df['class']]

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['odor'], y=df['class'], hue=df['class'], palette={'e': 'blue', 'p': 'orange'}, legend='full')

plt.xlabel('Odor')
plt.ylabel('Class')
plt.title('Odor as edible vs poisonous')

plt.show()
```

# Conclusion
